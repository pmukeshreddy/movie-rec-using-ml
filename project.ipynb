```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

```python
movies_data = pd.read_csv("movies.csv")
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
movies_data.isnull().sum()
```

```python
selected_features=["genres","keywords","tagline","cast","director"]
selected_features
```

```python
for feature in selected_features:
    movies_data[feature]=movies_data[feature].fillna(" ")
```

```python
combined_features=movies_data["genres"]+movies_data["keywords"]+movies_data["tagline"]+movies_data["cast"]+movies_data["director"]
```

```python
print(combined_features)
```

```python
combined_features.shape
```

```python
vectorizer=TfidfVectorizer()
```

```python
feaurted_vector = vectorizer.fit_transform(combined_features)
```

```python
feaurted_vector.shape
```

```python
print(feaurted_vector)
```

```python
print(feaurted_vector.shape)
```

cosine similarity

```python
similartiy = cosine_similarity(feaurted_vector)
```

```python
print(similartiy)
```

```python
similartiy.shape
```

```python
find_close_match = difflib.get_close_matches(movie_name,lis_name)
find_close_match
```

```python
movies_data[movies_data.title==find_close_match[0]]
```

```python
close_match = find_close_match[0]
close_match
```

```python
index_of_movie = movies_data[movies_data.title==close_match]["index"].values[0]
index_of_movie
```

```python
similartiy_score = list(enumerate(similartiy[index_of_movie]))
print(similartiy_score)
```

```python
len(similartiy_score)
```

```python
sorted_similar_movies = sorted(similartiy_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
print("movie's suggestion for you \n")
i = 1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index==index]["title"].values[0]
    if (i<30):
        print(i,".",title_from_index)
        i+=1
```

```python
movie_name = input(" Enter your favorite movie \n")

list_of_all_titile = movies_data["title"].tolist()

find_close_match = difflib.get_close_matches(movie_name,list_of_all_titile)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title==close_match]["index"].values[0]

similartiy_score = list(enumerate(similartiy[index_of_movie]))

sorted_similar_movies = sorted(similartiy_score,key=lambda x:x[1],reverse=True)

i = 1

for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index = movies_data[movies_data.index==index]["title"].values[0]
    if (i<30):
        print(i,".",title_from_index)
        i+=1
```
